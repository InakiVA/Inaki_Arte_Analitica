```python
import math
import pandas as pd
import os

def primos(n):
  primos=[]
  for i in range(2, n+1):
    isPrime= True
    limit=int(math.sqrt(i))
    for test in range(2,limit+1):
      if i % test ==0:
        isPrime=False
        break
    if isPrime:
      primos.append(i)
  return primos

```

```python
print(primos(100))
```

```python
def crossProductA(a,b):
  #a=(ax,ay,az)
  #b=(bx,by,bz)  
  product=[a[1]*b[2] - a[2]*b[1], a[2]*b[0] - a[0]*b[2], a[0]*b[1] - a[1]*b[0]]
  return product
```

```python
vectorA=[1,0,0]
vectorB=[0,1,0]
productoCruz=crossProductA(vectorA,vectorB)
print(productoCruz)
```
